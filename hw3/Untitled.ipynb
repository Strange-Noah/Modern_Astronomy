# Stellar Lifetimes

## (a) Main Sequence lifetime of the Sun

Energy released when 4 H fuse into 1 He:
- Mass converted to energy = 0.7% = 0.007

Using Einstein’s relation:

:contentReference[oaicite:0]{index=0}

Total energy available from hydrogen burning:
- Only 10% of the Sun’s mass participates in core fusion
- Energy = 0.1 M☉ × 0.007 × c² = 0.0007 M☉ c²

Using:
- M☉ ≈ 2 × 10³³ g  
- c ≈ 3 × 10¹⁰ cm/s  

Energy:
E ≈ 0.0007 × (2 × 10³³) × (3 × 10¹⁰)²  
≈ 1.26 × 10⁵¹ erg  

Lifetime:
t = E / L  
= (1.26 × 10⁵¹) / (3.8 × 10³³)  
≈ 3.3 × 10¹⁷ s  

Convert to years:
t ≈ 1.0 × 10¹⁰ years  

**Answer:** The Sun stays on the Main Sequence for about **10 billion years**.

---

## (b) Lifetime of a 10 M☉ star

Given:
- Luminosity relation: L ∝ M⁴  
- Lifetime: t ∝ M / L ∝ M / M⁴ = M⁻³  

Thus:
t / t☉ = (M / M☉)⁻³ = 10⁻³  

t ≈ 10¹⁰ × 10⁻³ = 10⁷ years  

**Answer:** A 10 M☉ star lives about **10 million years**.

---

## (c) Lifetime of a 0.3 M☉ star

t / t☉ = (0.3)⁻³ ≈ 37  

t ≈ 3.7 × 10¹¹ years  

Age of the universe ≈ 1.4 × 10¹⁰ years  

**Answer:**  
A 0.3 M☉ star lives about **370 billion years**, which is **much longer than the age of the universe**.

**Meaning:**  
No such low-mass star has yet left the Main Sequence; all are still burning hydrogen today.

---

## (d) Evolution of a brown dwarf

A brown dwarf cannot sustain hydrogen fusion.

As it ages:
- It slowly radiates away its internal heat
- Becomes cooler and dimmer over time
- Eventually fades into a cold, dark object

**Answer:**  
A brown dwarf **gradually cools and fades**, becoming increasingly faint without ever entering the Main Sequence.

# Searching for Extrasolar Planets

## (a) Five techniques to detect exoplanets

1. **Radial Velocity (Doppler Method)**  
   Measures periodic shifts in a star’s spectral lines due to the gravitational pull of an orbiting planet.

2. **Transit Method**  
   Detects the slight dimming of a star when a planet passes (transits) in front of it.

3. **Direct Imaging**  
   Takes actual images of planets by blocking out the star’s light.

4. **Gravitational Microlensing**  
   Uses the bending of light from a background star when a foreground star (with planets) passes in front.

5. **Astrometry**  
   Measures tiny positional shifts in a star’s location on the sky due to orbiting planets.

---

## (b) Meaning of bias (scientific context)

**Bias** refers to a systematic tendency of a method or observation to favor detecting certain types of objects over others.

It means:
- The data collected is **not a perfectly representative sample**
- Some objects are **easier to detect**, so they appear more common than they actually are

---

## (c) Biases of each method

1. **Radial Velocity**
   - Favors **massive planets**
   - Favors planets **close to the star** (short periods)
   - Hard to detect small, Earth-like planets

2. **Transit Method**
   - Requires **edge-on alignment**
   - Favors **large planets** (block more light)
   - Favors **close-in planets** (more frequent transits)

3. **Direct Imaging**
   - Favors **very large, young, hot planets**
   - Favors planets **far from the star**
   - Difficult for Earth-like planets near stars

4. **Gravitational Microlensing**
   - Sensitive to planets at **moderate distances**
   - Events are **rare and not repeatable**
   - Hard to follow up or confirm

5. **Astrometry**
   - Favors **massive planets far from the star**
   - Requires extremely **high precision measurements**
   - Less sensitive to small planets

---

## (d) Has an Earth twin been detected?

**Answer:**  
No confirmed true **Earth twin** has been detected yet.

### Why not?

An Earth twin must have:
- Earth-like **mass and size**
- Earth-like **orbit (1 AU)**
- Around a **Sun-like star**
- Possibly Earth-like **atmosphere**

These are hard to detect because:
- Earth-sized planets produce **very weak signals**
- Long orbital period (~1 year) → requires **long observation time**
- Signals are near the **limit of current technology**

### Implications

- We **have detected many Earth-sized planets** (e.g., in habitable zones)
- But not a perfect Earth analog yet
- Current data suggests **Earth-like planets may be common**, but:
  - Detection biases mean we **under-detect them**

# The Big Bang

## (a) Description of the Big Bang Theory

The **Big Bang Theory** states that:
- The universe began about **13.8 billion years ago**
- It started from an extremely **hot, dense state**
- Space itself has been **expanding** ever since

**Key idea:**  
The Big Bang was **not an explosion in space**, but an expansion **of space itself**.

**Where did it happen?**  
- It happened **everywhere in the universe at once**
- There is **no central point** in space where it occurred

---

## (b) Three strongest pieces of evidence

### 1. Expansion of the Universe (Hubble’s Law)
- Distant galaxies are moving away from us
- The farther they are, the faster they recede  
- Observed via **redshift** of light

**Implication:** The universe was once smaller and denser.

---

### 2. Cosmic Microwave Background (CMB)
- A faint, uniform radiation detected in all directions
- Remnant heat from the early universe (~380,000 years after the Big Bang)
- Temperature ≈ **2.7 K**

**Implication:** The universe was once **hot and dense**

---

### 3. Abundance of Light Elements
- Observed amounts of **hydrogen, helium, and lithium**
- Match predictions from **Big Bang nucleosynthesis**

**Implication:** These elements formed in the first few minutes after the Big Bang

---

## (c) Future of the Universe

### Current understanding:
- The universe will likely **continue expanding forever**
- Expansion is **accelerating**

### Why do we think this?

- Observations of distant **Type Ia supernovae**
- Evidence for **dark energy**, which drives acceleration

### Possible future scenario:

**Heat Death (most accepted):**
- Galaxies move farther apart
- Stars burn out
- Universe becomes **cold, dark, and dilute**

# The Expanding Universe

## (a) Importance of Hubble’s Law

Hubble–Lemaître Law:

:contentReference[oaicite:0]{index=0}

**Importance:**
- Shows the universe is **expanding**
- Provides evidence for the **Big Bang Theory**
- Allows estimation of the **age of the universe** (~1/H₀)
- Establishes a relationship between **distance and redshift**
- Forms the foundation of **modern observational cosmology**

---

## (b) Calculating Hubble’s Constant from Hubble (1929)

**Method:**
1. Take data from Hubble’s Table 1:
   - Distance \(d\)
   - Velocity \(v\)
2. Plot **v vs. d**
3. Fit a straight line → slope = \(H_0\)

**Result (historical):**
- Hubble obtained:
\[
H_0 \approx 500 \,\text{km s}^{-1}\text{Mpc}^{-1}
\]

---

## (c) Comparison with current value

**Modern value:**
- \(H_0 \approx 70 \,\text{km s}^{-1}\text{Mpc}^{-1}\)

### Comparison:
- Hubble’s value ≈ **7 times too large**

### On a plot:
- Hubble (1929): **steeper slope**
- Modern value: **shallower slope**

This means:
- For the same distance, Hubble predicted **much higher velocities**

---

## (d) Why was Hubble’s value incorrect?

Several key reasons:

### 1. Distance calibration errors
- Misidentified **Cepheid variable stars**
- Used incorrect luminosities → distances underestimated

### 2. Limited data
- Small sample size
- Nearby galaxies only

### 3. Poor instrumentation
- Less accurate telescopes and detectors
- Large observational uncertainties

### 4. No distinction between stellar populations
- Confusion between **Population I and II Cepheids**
- Led to systematic distance errors

### 5. Unaccounted effects
- Peculiar velocities (non-Hubble motion)
- Dust extinction not properly corrected

```python
import matplotlib.pyplot as plt
import numpy as np
from pathlib import Path

file_path = Path.cwd() / "hubble29.csv"

data = np.genfromtxt(file_path, delimiter=',', names=True, dtype=None, encoding='utf-8')

d = data['distance']
v = data['velocity']

H0_fit, b_fit = np.polyfit(d, v, 1)

H0_modern = 70

d_line = np.linspace(0, max(d)*1.1, 100)

v_fit = H0_fit * d_line + b_fit
v_modern = H0_modern * d_line

plt.figure(figsize=(8,6))

plt.scatter(d, v, label='Hubble (1929) Data')

plt.plot(d_line, v_fit, linestyle='--',
         label=f'Hubble Fit: v = {H0_fit:.1f} d + {b_fit:.1f}')

plt.plot(d_line, v_modern, linestyle='-',
         label=f'Modern Value: v = {H0_modern} d')

plt.xlabel('Distance (Mpc)')
plt.ylabel('Velocity (km/s)')
plt.title('Hubble (1929) Data and Modern Comparison')

plt.legend()
plt.grid()

plt.savefig('hubble_plot.png', dpi=300)

plt.show()

print(f"Hubble (1929) fitted H0 = {H0_fit:.2f} km/s/Mpc")
```

# Galaxy Types

## (a) What kind of galaxy is the Milky Way? How do we know?

The Milky Way is a **barred spiral galaxy**.

**Evidence:**
- Observations in **infrared** reveal a central bar structure
- Distribution of stars and gas shows **spiral arms**
- Rotation curve and stellar motions match spiral galaxy dynamics

---

## (b) Why do galaxies have different shapes? Why red vs blue?

### Shapes:
Galaxy shapes are determined by:
- **Initial conditions** (angular momentum, mass)
- **Mergers and interactions**
- **Star formation history**
- **Gas content**

Examples:
- Spiral galaxies: high angular momentum, lots of gas
- Elliptical galaxies: result of mergers, little gas

---

### Colors:

- **Blue galaxies:**
  - Actively forming stars
  - Contain many **hot, massive, short-lived stars**
  
- **Red galaxies:**
  - Little or no star formation
  - Dominated by **old, cool stars**

**Reason:**  
Massive blue stars evolve quickly → if star formation stops, galaxy becomes red over time.

---

## (c) Globular clusters

**Definition:**
- Dense, spherical collections of stars
- Found in the **halo** of galaxies

**Typical number of stars:**
- ~10⁵ to 10⁶ stars

**Milky Way comparison:**
- Milky Way: ~10¹¹ stars

**Conclusion:**
- A globular cluster has **~100,000 times fewer stars** than the Milky Way

---

## (d) Evidence for supermassive black holes

At least two key methods:

1. **Stellar orbits near the center**
   - Stars move at very high speeds around an unseen massive object
   - Example: stars orbiting the center of the Milky Way

2. **Active Galactic Nuclei (AGN) / quasars**
   - Extremely luminous centers powered by accretion onto a black hole

3. (Optional extra)
   - **X-ray emission** from hot accretion disks

---

## (e) Evidence for dark matter

At least two strong lines of evidence:

1. **Galaxy rotation curves**
   - Stars orbit faster than expected at large radii
   - Implies unseen mass

2. **Gravitational lensing**
   - Light bending reveals more mass than visible matter

3. (Optional extra)
   - **Galaxy cluster dynamics**
   - **Cosmic microwave background measurements**

---

## (f) JWST deep field image labeling

### (i) Elliptical galaxy
- Smooth, round or oval shape
- No visible spiral structure
- Uniform brightness

---

### (ii) Spiral galaxy
- Disk shape with **spiral arms**
- Often shows dust lanes and star-forming regions

---

### (iii) Gravitational lensing (at least two)
- **Arcs or distorted shapes**
- Light from background galaxies bent around massive foreground objects

---

### (iv) Early universe galaxy
- Very **redshifted** → appears red
- Small, faint, irregular shape

---

### (v) Quasar (early universe)
- Extremely bright point-like source
- Often brighter than host galaxy
- Strong emission across wavelengths

# Science in Your Life

## (a) Steps in the Scientific Method

1. **Observation**  
   Notice a phenomenon or problem.

2. **Question**  
   Ask a clear question about the observation.

3. **Hypothesis**  
   Propose a testable explanation or prediction.

4. **Experiment/Test**  
   Design and perform experiments to test the hypothesis.

5. **Analysis**  
   Examine the data and results.

6. **Conclusion**  
   Determine whether the hypothesis is supported or rejected.

7. **Iteration**  
   Refine the hypothesis and repeat if necessary.

---

## (b) Example from everyday life

**Example: Improving study efficiency**

1. **Observation:**  
   I feel tired and less focused when studying for long periods.

2. **Question:**  
   Does taking short breaks improve my concentration?

3. **Hypothesis:**  
   Taking a 5-minute break every 25 minutes will improve focus and productivity.

4. **Experiment:**  
   Study for several days using:
   - Day 1–2: no breaks  
   - Day 3–4: 25 min study + 5 min break (Pomodoro method)

5. **Analysis:**  
   Compare:
   - Amount of material covered  
   - Level of concentration  
   - Fatigue level  

6. **Conclusion:**  
   If performance improves with breaks, the hypothesis is supported.

7. **Iteration:**  
   Try different intervals (e.g., 50 min + 10 min) to optimize results.
