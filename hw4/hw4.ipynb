# 1. Seasons, Lunar Phases, and the Earth

## (a) What causes the seasons?

The seasons are caused by the tilt of the Earth's rotation axis, not by changes in the Earth-Sun distance. The Earth's axis is tilted by about 23.5° relative to its orbit around the Sun.

As the Earth orbits the Sun, different hemispheres receive sunlight at different angles during different times of the year. When the Northern Hemisphere is tilted toward the Sun, sunlight strikes it more directly and the days are longer, producing summer. At the same time, the Southern Hemisphere experiences winter. Six months later, the situation reverses.

The intensity of sunlight depends on the angle at which sunlight hits the surface. More direct sunlight concentrates energy into a smaller area and increases heating.

## (b) What causes the lunar phases?

Lunar phases are caused by the changing relative positions of the Sun, Earth, and Moon as the Moon orbits the Earth.

Half of the Moon is always illuminated by the Sun. However, from Earth we see different fractions of the illuminated half depending on the Moon's position in its orbit.

For example:

- New Moon: the illuminated side faces away from Earth.
- First Quarter: half of the visible Moon is illuminated.
- Full Moon: the illuminated side faces Earth completely.
- Third Quarter: again half of the visible Moon is illuminated.

The lunar phases are not caused by the Earth's shadow. The Earth's shadow only causes lunar eclipses.

One complete cycle of lunar phases takes about 29.5 days.

## (c) Do any planets have phases? If so, which ones?

Yes. All planets can show phases to some extent, but the phases are most noticeable for the inner planets, Mercury and Venus.

Because Mercury and Venus orbit closer to the Sun than Earth does, we can observe them showing crescent, quarter, and nearly full phases, similar to the Moon.

The outer planets also technically have phases, but since they are much farther away from Earth and usually appear mostly illuminated, their phases are difficult to observe.

The observation of the phases of Venus was important evidence supporting the heliocentric model of the Solar System.

## (d) Why are some solar eclipses total eclipses and others not total? What are lunar eclipses? Are solar eclipses or lunar eclipses more common?

A solar eclipse occurs when the Moon passes between the Earth and the Sun.

Some solar eclipses are total because the Moon appears large enough in the sky to completely cover the Sun. However, the Moon's orbit around Earth is elliptical, so the Earth-Moon distance changes over time.

- When the Moon is closer to Earth, its apparent size is larger, producing a total solar eclipse.
- When the Moon is farther from Earth, its apparent size is smaller than the Sun's, producing an annular eclipse in which a bright ring of sunlight remains visible.

A lunar eclipse occurs when the Earth passes between the Sun and the Moon, causing the Moon to move into the Earth's shadow. Lunar eclipses can only occur during a full moon.

Globally, solar and lunar eclipses occur with similar frequency. However, from a particular location on Earth, lunar eclipses are seen more often because they are visible from the entire nighttime side of Earth, while total solar eclipses are only visible along a narrow path.

The four sunset photos were taken between March and May. During this period, the sunset time became later and the location of the sunset gradually shifted northward along the horizon.

This change occurs because the Earth's axial tilt changes the Sun's apparent path across the sky throughout the year. As the Northern Hemisphere moves toward summer, the Sun appears higher in the sky and remains above the horizon for a longer time each day.

## (f) Is the dark side of the Moon ever visible from Earth?

The phrase "dark side of the Moon" is misleading. The far side of the Moon is not permanently dark. All parts of the Moon receive sunlight at different times.

The reason we usually see only one side of the Moon is tidal locking. The Moon rotates once on its axis in approximately the same amount of time that it takes to orbit the Earth. As a result, the same side of the Moon always faces Earth.

However, due to a small effect called libration, we can actually observe about 59% of the Moon's surface over time from Earth.

The far side of the Moon was first directly photographed by spacecraft, not from Earth.

## (h) The Moon is moving away from Earth at 3.8 cm/yr. Approximately what year will the last total solar eclipse occur?

A total solar eclipse occurs because the apparent sizes of the Moon and Sun in the sky are very similar.

As the Moon slowly moves away from Earth, its apparent angular size decreases. Eventually, the Moon will no longer appear large enough to completely cover the Sun, so total solar eclipses will no longer occur.

Currently:

- Average Earth-Moon distance: about \(3.84 \times 10^5\) km
- Recession rate: \(3.8\) cm/yr \(= 3.8 \times 10^{-5}\) km/yr

Estimates suggest that the Moon must move several tens of thousands of kilometers farther away before total eclipses become impossible. Using an approximate increase of \(2 \times 10^4\) km:

\[
t = \frac{2 \times 10^4 \text{ km}}{3.8 \times 10^{-5} \text{ km/yr}}
\]

\[
t \approx 5 \times 10^8 \text{ years}
\]

So the last total solar eclipse will likely occur roughly 500–600 million years from now.

## (i) The Earth's rotation is slowing by about 0.01 seconds each year. Would the Earth day ever equal the Earth year?

Currently:

- One Earth day is about \(86400\) seconds.
- One Earth year is about \(3.16 \times 10^7\) seconds.

The difference is:

\[
3.16 \times 10^7 - 86400 \approx 3.15 \times 10^7 \text{ seconds}
\]

If the length of the day increases by \(0.01\) seconds per year, then the approximate time required would be:

\[
t = \frac{3.15 \times 10^7}{0.01}
\]

\[
t \approx 3.15 \times 10^9 \text{ years}
\]

This is about 3 billion years.

In reality, this exact situation will probably never occur because the Sun will evolve significantly before then, and the Earth-Moon system will continue changing over time.

The slowing of Earth's rotation is caused mainly by tidal interactions with the Moon.

---

## (j) How are (h) and (i) related?

Parts (h) and (i) are consequences of the same physical process: tidal interactions between the Earth and the Moon.

The Moon's gravity creates tidal bulges on Earth. Because the Earth rotates faster than the Moon orbits Earth, the tidal bulges are carried slightly ahead of the Earth-Moon line.

This produces a gravitational torque that transfers angular momentum from Earth's rotation to the Moon's orbit.

As a result:

- The Earth's rotation gradually slows down, making the day longer.
- The Moon gains orbital energy and moves farther away from Earth.

Therefore:

- Part (h) occurs because the Moon is receding from Earth.
- Part (i) occurs because Earth's rotation is slowing down.

These are two connected consequences of the same tidal process.

Over very long timescales, tidal interactions tend to drive systems toward tidal locking, where rotational and orbital periods become synchronized.

# 2. Searching for Life in the Universe

## (a) What properties are most important for determining whether a planet is habitable?

A habitable planet is generally defined as a planet that can support liquid water on its surface for long periods of time, since liquid water is considered essential for life as we know it.

The most important properties for determining habitability include:

1. **Surface Temperature**
   - The temperature must allow liquid water to exist.
   - This depends on the distance from the host star, atmospheric composition, and greenhouse effects.

2. **Atmosphere**
   - A planet needs an atmosphere thick enough to maintain pressure and regulate temperature.
   - The atmosphere can also protect the surface from harmful radiation.

3. **Presence of Water**
   - Liquid water is one of the key requirements for life.
   - Water may exist on the surface or underground.

4. **Energy Source**
   - Life requires a continuous energy source, such as sunlight, geothermal activity, or chemical reactions.

5. **Magnetic Field**
   - A magnetic field can help protect the atmosphere from being stripped away by stellar winds.

6. **Chemical Composition**
   - Elements such as carbon, hydrogen, oxygen, nitrogen, phosphorus, and sulfur are important for known biological processes.

7. **Stability of the Host Star**
   - Highly variable stars may produce intense radiation that could make life difficult.

Overall, habitability means that environmental conditions are stable enough for complex chemistry and potentially life to exist.

## (b) What observations would help determine habitability? What about life?

To determine whether a planet is habitable, I would want observations of:

- The planet's temperature
- Atmospheric composition
- Presence of water vapor
- Surface pressure
- Orbital distance from the star
- Planet size and mass
- Evidence of clouds or oceans

Spectroscopy would be especially important because it allows us to study the atmosphere by analyzing absorption lines.

To search for life itself, I would look for possible biosignatures, including:

- Oxygen (\(O_2\))
- Ozone (\(O_3\))
- Methane (\(CH_4\))
- Water vapor (\(H_2O\))
- Organic molecules

The simultaneous presence of oxygen and methane could be especially important because these gases chemically react with each other and would need continuous replenishment, possibly by biological activity.

I would also look for seasonal atmospheric changes, unusual chemical disequilibrium, or surface reflectance features that might indicate vegetation.

---

## (c) Why is the search for life not currently feasible?

The search for life is extremely difficult because exoplanets are very far away and extremely faint compared to their host stars.

Some major challenges include:

1. **Small Signals**
   - The light from a planet is much weaker than the light from its star.
   - Detecting atmospheric features requires very precise measurements.

2. **Limited Spatial Resolution**
   - Most exoplanets cannot be directly imaged.
   - In many cases we only detect indirect signals such as transits or stellar wobble.

3. **Uncertain Biosignatures**
   - Many chemicals associated with life can also be produced by non-biological processes.
   - It is difficult to distinguish biological and geological sources.

4. **Lack of Detailed Data**
   - Current telescopes provide only limited atmospheric information for most exoplanets.

5. **Timescale Problem**
   - Even if intelligent life exists, civilizations may not overlap in time with humanity.

Future telescopes and missions may improve our ability to detect possible biosignatures.

## (d) Calculate the equilibrium temperature of Earth.

The equilibrium temperature is given by:

\[
T_{\mathrm{eq}}^4 =
\frac{L(1-a)}
{16\pi d^2 \sigma}
\]

Given:

\[
L = 3.83 \times 10^{26}\ \mathrm{W}

\[
a = 0.3
\]

\[
d = 1\ \mathrm{AU} = 1.496 \times 10^{11}\ \mathrm{m}
\]

\[
\sigma = 5.67 \times 10^{-8}\ \mathrm{W\,m^{-2}\,K^{-4}}
\]

Substituting:

\[
T_{\mathrm{eq}}^4 =
\frac{
(3.83 \times 10^{26})(1-0.3)
}
{
16\pi (1.496 \times 10^{11})^2
(5.67 \times 10^{-8})
}
\]

Evaluating:

\[
T_{\mathrm{eq}} \approx 255\ \mathrm{K}
\]

This is about:

\[
255\ \mathrm{K} \approx -18^\circ \mathrm{C}
\]

This is lower than Earth's actual average surface temperature because the greenhouse effect warms the Earth.

## (e) Distance from stars of different masses for Earth-like equilibrium temperature

For the same equilibrium temperature, the received stellar flux must remain constant.

Since:

\[
F \propto \frac{L}{d^2}
\]

and:

\[
L \propto M^4
\]

we obtain:

\[
d \propto \sqrt{L}
\propto M^2
\]

Therefore, compared with Earth:

### For a \(5M_\odot\) star:

\[
d = 5^2 = 25\ \mathrm{AU}
\]

So the planet would need to orbit at approximately:

\[
25\ \mathrm{AU}
\]

to have the same equilibrium temperature as Earth.

---

### For a \(0.2M_\odot\) star:

\[
d = (0.2)^2 = 0.04\ \mathrm{AU}
\]

So the planet would need to orbit at approximately:

\[
0.04\ \mathrm{AU}
\]

to have the same equilibrium temperature as Earth.

This shows that habitable zones around low-mass stars are much closer to the host star.

## (f) Are there places in our Solar System besides Earth where it makes sense to search for life?

Yes. Several locations in the Solar System are considered promising places to search for life.

### Mars

Mars may once have had liquid water on its surface. Evidence includes:

- Ancient river channels
- Mineral deposits formed in water
- Polar ice caps
- Possible underground water

Microbial life may potentially survive underground where conditions are more stable.

---

### Europa

Europa, a moon of Jupiter, is one of the best candidates for extraterrestrial life.

Evidence suggests that Europa has:

- A global subsurface ocean
- Internal heating from tidal forces
- Possible hydrothermal activity on the ocean floor

These conditions may resemble deep ocean environments on Earth where life exists without sunlight.

---

### Enceladus

Enceladus, a moon of Saturn, ejects water vapor and ice particles into space through geysers near its south pole.

Observations indicate:

- Liquid water beneath the surface
- Organic molecules
- Internal heating

This makes Enceladus another strong candidate for microbial life.

---

### Titan

Titan, another moon of Saturn, has a thick atmosphere and lakes of liquid methane and ethane.

Although conditions are very cold, Titan may support unusual forms of chemistry that could be relevant to life.

Overall, Europa and Enceladus are currently among the most promising places to search for extraterrestrial microbial life in our Solar System.
